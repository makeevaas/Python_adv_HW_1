```python
import pandas as pd
```

```python
#Анализ успеваемости студентов

train_data = pd.read_csv('train1.csv', sep = ',')

#1 Признак - вычислить среднее количество вопросов на которые были даны правильные ответы
#Получить из train.csv только все вопросы
only_questions = train_data[(train_data['content_type_id'] == 0)]
#Получить из train.csv только вопросы на которые были даны верные ответы
only_correct_questions = train_data[(train_data['content_type_id'] == 0) & (train_data['answered_correctly'] == 1)]

#Вычислить количество и процентное соотношение верных ответов из общего количества пройденных вопросов
count_only_questions = only_questions['row_id'].count()
count_only_correct_questions = only_correct_questions['row_id'].count()
in_percents = (count_only_correct_questions * 100) // count_only_questions

#Результат анализа
result = pd.DataFrame({
    '| all questions': [count_only_questions],
    '| correct answers': [count_only_correct_questions],
    '| % correct answers': [in_percents]
})

display(result)

print('из всех пройденных вопросов, на %s процентов даны верные ответы' %in_percents)
```

```python
#Анализ успеваемости студентов

train_data = pd.read_csv('train1.csv', sep = ',')

#2 Признак - вычислить среднее время прохождения всеми студентами курсов и сравнить с граничными порогами(минимальным/максимальным) - оценить тенденцию
# Получить максимальное, минимальное и среднее время прохождения курса от начала программы до текущего момента из train.csv
max_value = train_data['timestamp'].max()
min_value = train_data['timestamp'].min()
mean_value = train_data['timestamp'].mean()

#Получить количество данных ниже и выше стреднего
count_above_average = train_data[(train_data['timestamp'] >= mean_value) & (train_data['timestamp'] <= max_value)]['row_id'].count()
count_bellow_average = train_data[(train_data['timestamp'] >= min_value) & (train_data['timestamp'] <= mean_value)]['row_id'].count()

#Сравнить и вычислить в процентном соотношении время общей успеваемости, оценить тенденцию
#Результат анализа
result = pd.DataFrame({
    '| max time': [max_value],
    '| min time': [min_value],
    '| mean time': [mean_value],
    '| count above average': [count_above_average],
    '| count bellow average': [count_bellow_average]
    })

display(result)

if count_above_average > count_bellow_average:
  difference=count_above_average-count_bellow_average
  print('Тенденция успеваемости студентов по курсам на основании времени прохождения: Количество курсов, пройденных за время выше среднего, превышает количество курсов, пройденных за время ниже среднего, на %s' %difference)
if count_above_average < count_bellow_average:
  difference=count_bellow_average-count_above_average
  print('Тенденция успеваемости студентов по курсам на основании времени прохождения: Количество курсов, пройденных за время ниже среднего, превышает количество курсов, пройденных за время выше среднего, на %s' %difference)  
```

```python
#Анализ успеваемости студентов

train_data = pd.read_csv('train1.csv', sep = ',')
questions_data = pd.read_csv('questions.csv', sep = ',')
lectures_data = pd.read_csv('lectures.csv', sep = ',')

#3 Признак - получить наиболее эффективные лекции по цели на основе анализа данных
#Получить из train.csv только вопросы на которые были даны верные ответы
only_correct_questions = train_data[(train_data['content_type_id'] == 0) & (train_data['answered_correctly'] == 1)]

#Получить из questions.csv по id те вопросы на которые даны правильные ответы.
questions=only_correct_questions.merge(questions_data, left_on='content_id', right_on='question_id', how='inner')

#Собрать тэги лекций из таблицы questions и сгрупировать
tags_in_questions=questions[['tags']].groupby(['tags']).mean()


#Функция перебора тэгов, поиск 
def filt_tags(tags):
    #создать список уникальных тэгов
    unique_tags=[]
    for i,row in tags.iterrows():
      #Собрать не повторяющиеся тэги в список
        el=i.split()
        if len(el)>1:
          for t in el:
            if t not in unique_tags:
              unique_tags.append(t)
        else:
          if el[0] not in unique_tags:
            unique_tags.append(el[0])
    return unique_tags

#Создать список уникальных тэгов лекций           
lectures_tags=filt_tags(tags_in_questions)

#Функция для получения данных из таблицы lectures.csv по тэгам
def get_lectures(tags):
    count_concept=0
    count_solving_question=0
    #Перебрать список с тэгами
    for t in tags:
      #Получить из таблицы lectures.csv лекцию по тэгу
      lecture = lectures_data[(lectures_data['tag'] == int(t))]
      #Игнорировать пустые
      if not lecture.empty:
        for i,row in lecture.iterrows():
          purpose_lecture=row['type_of']
          #Собрать по количеству цели concept и olving question
          if purpose_lecture=='concept':
            count_concept+=1
          if purpose_lecture=='solving question':  
            count_solving_question+=1
    return count_concept, count_solving_question

#Получить количество лекций, посчитынным по категории цели    
count_concept_lecture,count_solving_question_lecture=get_lectures(lectures_tags)

#Результат анализа
result = pd.DataFrame({
    '| count concept lecture': [count_concept_lecture],
    '| count solving question lecture': [count_solving_question_lecture],
})

display(result)

print('Количество лекций из категории concept, на вопросы по которым были даны верные ответы составляет: %s' %count_concept_lecture)
print('Количество лекций из категории solving question, на вопросы по которым были даны верные ответы составляет: %s' %count_solving_question_lecture)
```
